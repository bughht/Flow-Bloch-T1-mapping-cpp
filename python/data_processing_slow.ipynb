```python
MOLLI_readout_t = np.array( [200, 1200, 2200, 3200, 4200, 300, 1300, 2300]) * 1e-3
t_fit = np.arange(0, 5, 1e-4)
T1_list = []
T1_star_list = []
A_list = []
B_list = []
plt.figure(figsize=(10,5))
for v_,molli_ in zip(v,molli):
    T1_star, T1, A, B=MF_MAGIR(MOLLI_readout_t,molli_)
    T1_list.append(T1)
    T1_star_list.append(T1_star)
    A_list.append(A)
    B_list.append(B)
    plt.plot(t_fit,T1_fit_abs(t_fit, T1_star, A, B),alpha=0.5,label=str(v_))
    plt.plot(MOLLI_readout_t,molli_,"x")
    # plt.legend()

plt.show()
# T1_list
```

```python
plt.scatter(v,T1_list,label="T1")
plt.scatter(v,T1_star_list,label="T1*")
# plt.scatter(v,np.array(B_list)/np.array(A_list))
# plt.title("T1 = 1600 ms")
plt.ylabel("T1 (s)")
plt.xlabel("v (m/s)")
plt.legend()
plt.show()
# plt.plot(v,A_list,label="A")
# plt.plot(v,B_list,label="B")
plt.scatter(v,A_list,label="A")
plt.scatter(v,B_list,label="B")
plt.ylabel("A and B (params for T1 fit)")
plt.xlabel("v (m/s)")
plt.legend()
plt.show()
```
